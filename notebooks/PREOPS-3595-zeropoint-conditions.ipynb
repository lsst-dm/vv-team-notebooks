## Obtaining the zeropoint, airmass, PSF, and camera information

We obviously need zeropoints, airmasses, and PSFs. Camera information -- specifically the position of each detector in the focal plane -- is needed because, as we'll see, a frame's zeropoint has a strong dependence on the position of the detector in the focal plane. The zeropoint, airmass, and PSFs are taken from the finalVisitSummary table, which is post-FGCM.

From the main repo, work with the HSC data processed with the wk. 23 version of the pipeline:

```python
repo = '/repo/main'
collections = ['HSC/runs/PDR2/v24_1_0/DM-39132/step7']
butler = dafButler.Butler(repo, collections=collections)
registry = butler.registry
```

Get the camera information from the butler:

```python
camera = butler.get('camera', dataId={'instrument':'HSC'})
```

And query all visits conducted by the HSC with airmass < 2 (i.e., zenith angle < 60 deg), and report how many there are: 

```python
cVT_ref = list(registry.queryDatasets('ccdVisitTable', where="instrument='HSC'"))
cVT = butler.get(cVT_ref[0])
```

```python
cVT_sub = cVT[['visitId',
               'band',
               'zenithDistance',
               'zeroPoint',
               'psfSigma',
               'detector',
               'seeing',
               'expTime']]
cVT_sub = cVT_sub.reindex(columns = cVT_sub.columns.tolist() + ['xpos','ypos','airmass'])
```

Use the first dataId to get a list of detectors, then refer to the "camera" dataset to obtain the position of each detector one the focal plane.

```python
detectorList = list(set(cVT_sub['detector']))
detectorList = np.sort(detectorList)
xposList = np.array([camera[detector].getOrientation().getFpPosition().getX() for detector in detectorList])
yposList = np.array([camera[detector].getOrientation().getFpPosition().getY() for detector in detectorList])
for detector, xpos, ypos in zip(detectorList, xposList, yposList):
    mask = cVT_sub['detector'] == detector
    cVT_sub.loc[mask, 'xpos'] = xpos
    cVT_sub.loc[mask, 'ypos'] = ypos
```

```python
zenithDist = cVT_sub['zenithDistance'] * np.pi / 180.
cVT_sub['airmass'] = 1. / np.cos(zenithDist)
```

```python
cVT_sub['zeroPoint'] = cVT_sub['zeroPoint'] - 2.5 * np.log10(cVT_sub['expTime'])
```

```python
cVT_sub.to_csv('HSC_airmass_magzero_fgcm.csv')
```

### Fit the positional dependency of the zeropoint
The following is a function that will fit a 3D parabolic function.

```python
def fit2d(x, y, z):
    poly = PolynomialFeatures(degree=2)

    input_pts = np.stack([x, y]).T
    in_features = poly.fit_transform(input_pts)

    reg = LinearRegression().fit(in_features, z)
    
    score = reg.score(in_features, z)
    
    pred = reg.predict(in_features)
    
    return (reg, score, pred)
```

Perform the fit and subtract the positional dependency. Add the positionally-corrected zeropoint to the dataframe as "corrected_zp" column. Also create a dataframe to store the fit coefficients and score. Write the new dataframes to csv.

```python
df['corrected_zp'] = 0.

visits = np.array(list(set(df['visitId'])))

coeff_x2 = np.zeros(len(visits))
coeff_y2 = np.zeros(len(visits))
score = np.zeros(len(visits))
intercept = np.zeros(len(visits))
airmass = np.zeros(len(visits))
psfSigma = np.zeros(len(visits))

band = []

for i, visit in enumerate(visits):
    index = df['visitId'] == visit
    selected = df[index]
    boresight = selected.query('detector == 49')
    
    reg, score[i], pred = fit2d(selected['xpos'] / 200.,
                                selected['ypos'] / 200.,
                                selected['zeroPoint'])
    
    df.loc[index,'corrected_zp'] = df.loc[index,'zeroPoint'] - pred + reg.intercept_
    
    band.append(selected['band'].values[0])
    coeff_x2[i] = reg.coef_[3]
    coeff_y2[i] = reg.coef_[5]
    intercept[i] = reg.intercept_
    
    #airmass[i] = boresight['airmass'].values
    #psfSigma[i] = boresight['psfSigma'].values
    
band = np.array(band)
fitResults = {'visit':visits,
              'band':band,
              'coeff_x2':coeff_x2,
              'coeff_y2':coeff_y2,
              'score':score}
fR = pd.DataFrame(fitResults)
fR.to_csv('HSC_zeropoint_xyFitResults.csv')

df.to_csv('HSC_airmass_magzero_fgcm_xyCorrected.csv')
```
